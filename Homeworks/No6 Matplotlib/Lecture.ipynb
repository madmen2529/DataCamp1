```python
import numpy as np
# คำสั่งมาจาก pyplot module
import matplotlib.pyplot as plt
# ทำให้มองเห็น plot ใน Jupyter notebook
%matplotlib inline
```

```python
#Plt.plot(x,y)
#เป็นคำสั่งให้พล็อตกราฟ
x = [1,2,3,4,5,6,7,8,9]
x = np.array(x)
y = x**2
```

```python
plt.plot(x,y)
# ถ้าไม่ได้ใช้ Jupyter notebook
# เพื่อแสดงผล
plt.show()
```

```python
#Plt.xlabel( )
#Plt.xlabel( )
#Plt.title( )

#เป็นคำสั่งให้ใส่ชื่อแกน และ ชื่อกราฟ
#Plt.show() อยู่ล่างสุด

plt.plot(x,y)
# ถ้าไม่ได้ใช้ Jupyter notebook
# plt.show() เพื่อแสดงผล
plt.xlabel('1-9')
plt.ylabel('1-9 squared')
plt.title('Title is here')
plt.show()
```

```python
x = np.arange(0,90)
y = np.sin(x)
plt.plot(x,y)
```

```python
#Plt.plot(x,y,ls=‘ ’)
#เป็นคำสั่งให้เปลี่ยนเส้น Line Style
x = np.arange(0,20)
y = np.sin(x)
plt.plot(x,y,ls='--')
plt.title('Sin')
plt.show()
```

```python
#Plt.plot(x,y1)
#Plt.plot(x,y2)

#สร้างกราฟซ้อนกัน
#สามารถซ้อนได้มากกว่า2

x = np.arange(0,20)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#Label เพื่อระบุว่า
#เส้นนี้คืออะไร  
x = np.arange(0,30)
plt.plot(x,x**2,label='x**2 function')
plt.title('Trigonometry')
plt.plot(x,2*x**2,label='2x**2 function')
plt.legend(loc=0)
plt.show()
```

```python
#Plt.legend(loc=?) 
#เพื่อระบุตำแหน่งของ legend
x = np.arange(0,30)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,label='Sin function')
plt.title('Trigonometry')
plt.legend(loc=3)
plt.show()
```
