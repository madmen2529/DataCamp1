```python
import numpy as np
# คำสั่งมาจาก pyplot module
import matplotlib.pyplot as plt
# ทำให้มองเห็น plot ใน Jupyter notebook
%matplotlib inline
```

```python
#Plt.plot(x,y)
#เป็นคำสั่งให้พล็อตกราฟ
x = [1,2,3,4,5,6,7,8,9]
x = np.array(x)
y = x**2
```

```python
plt.plot(x,y)
# ถ้าไม่ได้ใช้ Jupyter notebook
# เพื่อแสดงผล
plt.show()
```

```python
#Plt.xlabel( )
#Plt.xlabel( )
#Plt.title( )

#เป็นคำสั่งให้ใส่ชื่อแกน และ ชื่อกราฟ
#Plt.show() อยู่ล่างสุด

plt.plot(x,y)
# ถ้าไม่ได้ใช้ Jupyter notebook
# plt.show() เพื่อแสดงผล
plt.xlabel('1-9')
plt.ylabel('1-9 squared')
plt.title('Title is here')
plt.show()
```

```python
#ใช้ ax เพื่อแยกกราฟออกจากกัน โดยคำสั่ง add_subplot และเลข 3 ตัว
#1 – อัตราส่วนแกน Y
#2 – อัตราส่วนแกน X
#3 – รูปที่
x = np.arange(0,20)
y = np.sin(x)
fig = plt.figure()
ax1 = fig.add_subplot(231,title='Num 1')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.plot(x,y,'m')

ax2 = fig.add_subplot(232, title='Num 2')
ax2.plot(y,x,'r')

plt.show()
```

```python
x = np.arange(0,30)
y = np.cos(x)
fig = plt.figure()
ax1 = fig.add_subplot(231, title='1/6')
ax1.plot(x,y,'r')
ax2 = fig.add_subplot(232, title='2/6')
ax2.plot(x,y,'b')
ax3 = fig.add_subplot(233, title='3/6')
ax3.plot(x,y,'m')
ax3.plot(x,y,'r')
ax4 = fig.add_subplot(234, title='4/6')
ax4.plot(x*3,y*3,'b')
ax5 = fig.add_subplot(235, title='5/6')
ax5.plot(x,np.cos(x+1),'p')
ax6 = fig.add_subplot(236, title='6/6')
ax6.plot(x,np.cosh(x),'m')
plt.show()
```

```python
#Plt.tight_layout() 
#เพื่อจัดกราฟให้เป็นระเบียบ
x = np.arange(0,30)
y = np.cos(x)
fig = plt.figure()
ax1 = fig.add_subplot(231, title='1/6')
ax1.plot(x,y,'r')
ax2 = fig.add_subplot(232, title='2/6')
ax2.plot(x,y,'b')
ax3 = fig.add_subplot(233, title='3/6')
ax3.plot(x,y,'m')
ax3.plot(x,y,'r')
ax4 = fig.add_subplot(234, title='4/6')
ax4.plot(x*3,y*3,'b')
ax5 = fig.add_subplot(235, title='5/6')
ax5.plot(x,np.cos(x+1),'p')
ax6 = fig.add_subplot(236, title='6/6')
ax6.plot(x,np.cosh(x),'m')
plt.tight_layout()
```

```python
#สามารถใช้ fig.add_axes เพื่อแทรกกราฟลงในกราฟใหญ่
#โดยแต่ละตัวเลขคือ
#1 – % จากซ้าย
#2 – % จากล่าง
#3 – % ความยาว
#4 -  % ความสูง
#จากจุด (0,0)
x = np.arange(0,30)
y = np.cos(x)
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.95,0.85])
ax1.plot(x,x**2,'r')
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(y,x,'b')
```

```python
#ใช้คำสั่ง plt.scatter()
#ใช้เพื่อดูตัวแปรนึงมีผลต่ออีกตัวแปรนึงอย่างไร
x = np.arange(1,11)
y = np.random.randint(1,150,10)
plt.figure(figsize=[5,5])
plt.scatter(x,y,c='r')
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.arange(1,11)
y = np.random.randint(1,150,10)
y2 = np.random.randint(10,110,10)
plt.figure(figsize=[5,5])
plt.scatter(x,y,c='r')
plt.scatter(x,y2,c='b')
plt.title('Scatter Plot')
plt.show()
```

```python
#ใช้คำสั่ง plt.bar()
#ใช้เพื่อดูปริมาณ เช่นยอดขาย
x = np.arange(1,11)
y = np.random.randint(1,150,10)
y = np.sort(y)
plt.figure(figsize=[5,5])
plt.bar(x,y)
plt.title('Bar Plot')
plt.show()
```

```python
#ใช้คำสั่ง plt.hist()
#คือกราฟแท่งแบบเฉพาะที่แสดงความสัมพันธ์ระหว่างข้อมูลเป็นหมวดหมู่ที่เรียกว่าชั้นข้อมูลกับความถี่ของข้อมูล เพื่อดูการกระจายของข้อมูล
x = np.random.randint(1,150,100)
plt.hist(x,bins=10)
plt.show()
```

```python
x = np.random.randint(1,3150,500)
plt.hist(x,bins=30)
plt.show()
```

```python
#ใช้คำสั่ง plt.pie()
#ใช้เปรียบเทียบสัดส่วนต่างๆเช่นรายได้ต่อเดือน รายจ่าย ยอดขาย
sales = np.random.randint(100,500,5)
plt.figure(figsize=[5,5])
plt.pie(sales)
plt.show()
```

```python
plt.figure(figsize=[5,5])
labels = ['Honda', 'Toyota', 'Mazda', 'Benz']
plt.pie([2,7,5,3], labels=labels)
plt.show()
```

```python
#สามารถใช้คำสั่ง explode เพื่อกำหนดส่วนที่ยื่นออกมาได้
plt.figure(figsize=[5,5])
explode = [0.1,0,0,0]
labels = ['Honda', 'Toyota', 'Mazda', 'Benz']
plt.pie([2,7,5,3], labels=labels,explode=explode)
plt.show()
```
