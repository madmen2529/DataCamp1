```python
#สำรวจข้อมูลที่หายไป ข้อมูลที่หายไปถูกบันทึกโดย แทน ? ด้วย ค่า NaN 
df = pd.read_csv('Automobile_data.csv', na_values={
    'normalized-losses':["?"],
    'stroke':["?"],
    'horsepower':["?"],
    'peak-rpm':["?"],
    'average-mileage':["?"]
})
df
```

```python
#ลบColumnข้อมูลที่หายไป
#Df.dropna(axis=1) 
#ลบrowข้อมูลที่หายไป
#Df.dropna()
df = pd.read_csv('Automobile_data.csv', na_values={
    'normalized-losses':["?"],
    'stroke':["?"],
    'horsepower':["?"],
    'peak-rpm':["?"],
    'average-mileage':["?"]
})
df.dropna(axis=1)
df.dropna()
df
```

```python
#เติมข้อมูลที่หายไป
#แทน ค่า NaN ด้วย 0
#df[‘column’].fillna(value=0)
df.fillna(value=0)
```

```python
#เติมข้อมูลที่หายไป

#แทน ค่า NaN ด้วย ค่าเฉลี่ยของcolumnนั้น เพื่อไม่ให้ค่าเฉลี่ยเปลี่ยน
#avg = df[‘column’].mean()

#df[‘column’].fillna(value=avg)
avg = df['normalized-losses'].mean()
print(avg)
df['normalized-losses'].fillna(avg, inplace=True)
```
